## Observations:

1. Max temperature's are observed around +/- 20 degree latitude's.

2. Both cloud cover and humidity does not seem to have a strong correlation to latitude.

3. Wind speeds appear to be slightly higher as we move away from the equator and max wind speeds are recorded around +/- 60 degress latitude.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint
import openweathermapy.core as owm
from urllib.error import HTTPError

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs=[]
cities=[]

lats = np.random.uniform(low=-90,high=90,size=1500)
lngs = np.random.uniform(low=-180,high=180,size=1500)

lat_lngs=zip(lats,lngs)
#print(list(lat_lngs))

for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0],lat_lng[1]).city_name
    
    if city not in cities:
        cities.append(city)
df = pd.DataFrame({"City":cities})
df.head()
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it's being processed (with the city number and city name).


```python
# List to store weather Data
weather_data=[]

# Creating settings dictionary to query keys and values 
settings={"units":"imperial","appid":api_key}

print("-----------------------------------------")
print("Beginning Data Retrieval")
print("-----------------------------------------")
j=1
i=1
# Looping through to cities List to make API calls
for city in cities:
    if(j<50):
        try:
            print(f"Processing Record {j} of Set {i}| {city})")
            weather_data.append(owm.get_current(city,**settings))
        except HTTPError:
            print(f"{city} not found... Skipping!")
        j+=1
    else:
        j=1
        i+=1
    time.sleep(.1)
print("-----------------------------------------")
print("Beginning Data Complete")
print("-----------------------------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#List of values to be extracted
data_extract=["name","clouds.all","sys.country","dt","main.humidity","coord.lat","coord.lon", "main.temp_max","wind.speed","main.temp"]

#Extracting required data
required_weather_data=[response(*data_extract) for response in weather_data]

#Defining column names for the DataFrame
column_names=["City","Cloudiness","Country","Date","Humidity","Lat","Lng","Max Temp","Wind Speed","Current Temp"]

#Building the DataFrame
df_weather_data=pd.DataFrame(required_weather_data,columns=column_names)

#Writing to a CSV file
df_weather_data.to_csv(output_data_file, index=False,header=True)

#Previewing DataFrame
df_weather_data.head()
```

### Plotting the Data

#### Latitude vs. Temperature Plot

```python
# First scatter plot: Latitude vs Max Temperature
date = time.strftime("%m/%d/%y", time.localtime(df_weather_data["Date"][0]))
plt.scatter(x=df_weather_data["Lat"],y=df_weather_data["Max Temp"], edgecolors="black",facecolor="lightblue")
plt.grid()
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.title(f"City Latitude vs. Max Temperature ({date})")
```
